```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
im = plt.imread('../data/example_image/ISIC_0001769.jpg')
plt.imshow(im);
```

```python
print(im.shape)
print(im.dtype)
```

```python
# Get a single RGB value from the blue circle (marker used by dermatologist)
print(im[2000,2000,:])
```

```python
#plotting all the colors
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(5, 3))
axes[0].imshow(im[:,:,0], cmap="gray")
axes[1].imshow(im[:,:,1], cmap="gray")
axes[2].imshow(im[:,:,2], cmap="gray")
fig.tight_layout()
```
